```python
''' Classificação automática por código do produto.
Caso produto comece por PER ele é perecível, Caso comece com PEN, não é perecível 
Exemplos de códigos: PER1235 e PEN9845'''

codigo = input("Enre com o código do produto: ").upper()
print('PER' in codigo)
print('Comida perecível para resposta', '\033[1m' + 'TRUE' + '\033[0m')
print('Comida', '\033[1m' + 'não' + '\033[0m', 'perecível para resposta', '\033[1m' + 'FALSE' + '\033[0m')
```

```python
"""Criar calculadora financeira com as entradas: FATURAMENTO BRUTO, TAXA DE IMPOSTOS(%) E CUSTO DE OPERAÇÃO.
Retorna o faturamento líquido """

fat_bruto = float(input("Entre com o faturamento bruto da empresa: "))
tax_imposto = float(input("Entre com a taxa de juros em %: "))
custo = float(input("Entre com o custo de operação da empresa: "))


print(f'O faturamento líquido da empresa foi de R$ {round(fat_bruto - fat_bruto * tax_imposto / 100 - custo, 2)}')


# OU

'''
fat_liquido = round(fat_bruto - fat_bruto * tax_imposto / 100 - custo, 2)

print(f'O faturamento líquido da empresa foi de R$ {fat_liquido}')
'''
```

```python
# Calcular a hipotenusa de um triângulo retângulo qualquer.

cateto_adj = float(input("Entre com a medida do cateto adjacente: "))
cateto_opo = float(input("Entre com a medida do cateto oposto: "))

print(f'A hipotenusa destes triângulo retângulo é: {(cateto_adj ** 2 + cateto_opo ** 2) ** (1/2)}')
```

```python
# Recebe segundos e retorna horas, minutos e segundos 

segundos_inicio = int(input('Entre com muito tempo em segundos: '))

minutos = segundos_inicio // 60
horas  = segundos_inicio // 3600

segundos_fim = int((segundos_inicio / 60 - minutos) * 60)

minutos_finais = int((minutos / 60 - horas) * 60)

print(f'{segundos_inicio} segundos equivalem a {horas} horas ou {minutos_finais} minutos e {segundos_fim} segundos ')
```

```python
"""
Calculadora do bar:
Informar: CONTA, PESSOAS GORJETA (%)
retorna: Total por pessoa, total da conta e o valor da gorjeta.
""" 
 
conta = float(input('Entre com o valor da conta: '))
gorjeta = int(input('Entre com a % de gorjeta (5, 10 ou 15): '))
pessoas = int(input('Entre com o número de pessoas que pagarão a conta: '))

total = conta + conta * gorjeta / 100
total_pessoal = round(total / pessoas, 2)

print(f'Cada pessoa pagará {total_pessoal}') 
```
