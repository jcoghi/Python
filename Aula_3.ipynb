# Estrutura de Repetição **FOR**

```python
# Imprimindo números de 0 a 9
for a in range (10):
    print(a) 
```

```python
# Imprimindo números de 1 a 10
for a in range (1, 11):
    print(a) 
```

```python
# Imprimindo os números pares
for a in range(0, 10, 2):
    print(a)
```

```python
# Buscando dentro de um texto
for x in 'banana':
    print(x)
```

```python
# Desempacotando uma lista
adj = ["red", "big", "tasty"]
fruits = ["apple", "banana", "cherry"]

for x in adj:
    for y in fruits:
        if x == 'red' and y == 'apple':
            print(x, y)
        elif x == 'big' and y == 'cherry':
            print(x, y)
        elif x == 'tasty' and y == 'banana':
            print(x, y)
```
