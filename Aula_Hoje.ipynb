```python
'''
Conversor de temperatura:
Entra temp_celsius e converte para temp_kelvin
temp_kelvin = 273.15 + temp_celsius
'''
# ENTRAR DADOS: input('Texto para o usuário saber o que tem que digitar')
# temp_c -> variavel -> tipo -> FLOAT -> Número decimal
# SAÍDA DE DADOS:print('Texto')/print(variavel) -> impressão na tela
# OUTRA FORMA: print('Começar o texto e, depois a {}'.format(variavel))

temp_celsius = input('Digite uma temperatura em °C: ')
temp_celsius = temp_celsius.replace(',', '.') # .replace (entrada, saída)
temp_kelvin = round(273.15 + float(temp_celsius), 2)

# round(conta, n° casa decimais) -> arredondamento

temp_kelvin = str(temp_kelvin)
temp_kelvin = temp_kelvin.replace('.', ',')


print(f'A temperatura em Kelvin é: {temp_kelvin}')
print('A temperatura em Kelvin é: {}'.format(temp_kelvin))
```

```python
"""Criar calculadora financeira com as entradas: FATURAMENTO BRUTO, TAXA DE IMPOSTOS(%) E CUSTO DE OPERAÇÃO.
Retorna o faturamento líquido """

# TODO
# ENTRAR: fat_bruto; taxa_imposto; custos OK
# imposto = fat_bruto * taxa_imposto/100
# CÁLCULO: fat_liquido = fat_bruto - imposto - custos
# RETORNAR: fat_liquido 

# ENTRADA
fat_bruto = float(input('Entre com o faturamento bruto: '))
taxa_imposto =float(input('Entre com a % da taxa de imposto: '))
custo =float(input('Entre com o custo: '))

#CONTAS
imposto = fat_bruto * taxa_imposto/100
fat_liquido = fat_bruto - imposto - custo

#SAÍDA
print('O faturamento líquido foi de R${:.2f}'.format(fat_liquido))

print('Faturamento bruto {}, Imposto {}, Custo {}, Líquido {}'.format(fat_bruto, imposto, custo, fat_liquido))
```

```python
# Calcular a hipotenusa de um triângulo retângulo qualquer.

# TODO
# ENTRADA: cateto a e b
# CONTA: (a ** 2 + b ** 2) ** (1/2)
#SAÍDA: informar qual é a hipotenusa

#ENTRADA
cateto_a = input('Entre com o cateto adjacente: ')
cateto_a = cateto_a.replace(',', '.')
cateto_b = input('Entre com o cateto oposto: ')
cateto_b = cateto_b.replace(',', '.')
# CONTA
hipotenusa = round((float(cateto_a) ** 2 + float(cateto_b) ** 2) ** (1/2),2)
hipotenusa = str(hipotenusa)
hipotenusa = hipotenusa.replace('.', ',')
#SAÍDA
print('A hipotenusa é {}'.format(hipotenusa))
```

```python
# Recebe segundos e retorna horas, minutos e segundos 
# ENTRADA: segundos -> inteiro
# horas = segundos / 3600 -> float
# minutos segundos / 60 -> float
# SAÍDA: X segundos equivalem a Y horas e a Z minutos

segundos = int(input('Entre com os segundos: '))
minutos = segundos / 60
horas = int(segundos / 3600)

segundos_restantes = int((segundos / 60 - int(minutos)) * 60)
minutos_restantes = int((minutos / 60 - horas) * 60)

print(f'{segundos} segundos equvalem a {horas} horas e {minutos_restantes} minutos e {segundos_restantes} segundos.')
```
