# Condicionais

As condicionais terão um papel determinante na nossa programação. Elas que farão as verificações necessárias para garantir que o programa atenda todas as possibilidades.

**Sua estrutura é:**
---
if condição:<br>
 > ação caso condição satisfeita<br>
---
Quando temos mais consições:
---
if condição:<br>
 > ação caso condição satisfeita <br>

elif condição:<br>
 > ação caso condição satisfeita <br>

else: <br>
 > ação caso condição satisfeita<br>
---

```python
# 1 - Adivinação de palavra

segredo = "cerveja"

adivinha = input("""Descubra sobre o que eu estou pensando:
É líquido, gostoso quando bem gelado \n""").casefold()

if adivinha == segredo:
  print("Acertou miseravi")
else:
  print('ERRRROUUUU!!!')
```

```python
''' Classificação automática por código do produto.
Caso produto comece por PER ele é perecível, Caso comece com PEN, não é perecível 
Exemplos de códigos: PER1235 e PEN9845'''

codigo = input("Enre com o código do produto: ").upper()
if ('PER' in codigo):
    print('Comida perecível para resposta')
else:
    print('Comida', '\033[1m' + 'não' + '\033[0m', 'perecível para resposta')
```

```python
"""Criar calculadora financeira com as entradas: FATURAMENTO BRUTO, TAXA DE IMPOSTOS(%) E CUSTO DE OPERAÇÃO.
Retorna o faturamento líquido """

fat_bruto = float(input("Entre com o faturamento bruto da empresa: "))
tax_imposto = float(input("Entre com a taxa de juros em %: "))
custo = float(input("Entre com o custo de operação da empresa: "))

fat_liquido = round(fat_bruto - fat_bruto * tax_imposto / 100 - custo, 2)

if fat_liquido > 0:
    print(f'O lucro da empresa foi de R$ {fat_liquido}')
elif fat_liquido <0:
    print(f'O prejuízo da empresa foi de R$ {fat_liquido}')
else:
    print('A empresa não teve lucro ou prejuízo.')
```
