# Function Python

### O que é?

As functions são blocos de código que servem 1 único propósito, fazem uma ação específica.

### Estrutura Básica


- Exemplo: vamos criar uma função de cadastro de um Produto. Essa função deve garantir que o produto cadastrado está em letra minúscula.

```python
def cadastrar_produto():
    produto = input("Entre com o nome do produto: ").casefold().capitalize().strip()
    print(f'{produto} foi cadastrado com sucesso.')
    # print(produto)
    return produto

lista = []

for i in range(3):
    lista.append(cadastrar_produto())

print(lista)
```
