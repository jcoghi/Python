```python
def numeros():
  x = input('Entre com o 1° número: ')
  x = x.replace(',','.')
  y = input('Entre com o 2° número: ')
  y = y.replace(',','.')
  return x, y

def soma():
  x, y = numeros()
  return float(x) + float(y)

resultado = soma()
print(resultado)
```

```python
def data_input():
  numbers_list = []
  count = 0
  while True:
    number = input(f'Entre com o {count+1}° número: ')
    number = number.replace(',','.')
    numbers_list.append(int(number))
    count += 1
    test = True
    while test:
        choice = input('Do you want to add another number? [Y/N] ').casefold()
        if choice == 'y':
            test = False
        elif choice == 'n':
          return print(f'The sum between {numbers_list} is {sum(numbers_list)}')
        else:
            print('Error, please enter Y for "Yes" or N for "No" ')


data_input()
```
