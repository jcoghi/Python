## Fun√ß√£o
Fun√ß√£o √© um trecho de c√≥digo respons√°vel por executar alguma tarefa espec√≠fica.

````python
# Vamos organizar nosso programa em fnu√ß√µes: Soma, Subtracao, Multiplicacao e Divis√£o
'''
Nosso programa deve trabalhar com as fun√ß√µes agora.

Crie as fun√ß√µes citadas. Dentro destas fun√ß√µes, voc√™ deve colocar tudo o que 
fez para calcular e printar as opera√ß√µes.

Dentro das condi√ß√µes do if, voc√™ deve chamar a fun√ß√£o para que ela funcione.
'''






````

````python
num1 = input('Entre com o 1¬∞ n√∫mero: ')
num1 = num1.replace(',','.')
num2 = input('Entre com o 2¬∞ n√∫mero: ')
num2 = num2.replace(',','.')
opcao= input('''Entre com o n√∫mero da op√ß√£o desejada:
1. Soma
2. Subtra√ß√£o
3. Multiplica√ß√£o
4. Divis√£o
''')

if opcao == '1':
    soma = float(num1) + float(num2)
    soma = str(soma).replace(".",",")
    print('A soma √© {}'.format(soma))
elif opcao == '2':
    subtracao = float(num1) - float(num2)
    subtracao = str(subtracao).replace(".",",")
    print('A subtra√ß√£o √© {}'.format(subtracao))
elif opcao == '3':
    multiplicacao = float(num1) * float(num2)
    multiplicacao = str(multiplicacao).replace(".",",")
    print('A multiplica√ß√£o √© {}'.format(multiplicacao))
elif opcao == '4':
    divisao = float(num1) / float(num2)
    divisao = str(divisao).replace(".",",")
    print('A divis√£o √© {}'.format(divisao))
else:
    print('Erro, selecione uma op√ß√£o v√°lida!')
````

````python
def Soma(num1, num2):
    soma = float(num1) + float(num2)
    soma = str(soma).replace(".",",")
    print('A soma √© {}'.format(soma))

def Subtracao(num1, num2):
    subtracao = float(num1) - float(num2)
    subtracao = str(subtracao).replace(".",",")
    print('A subtra√ß√£o √© {}'.format(subtracao))

def Multiplicacao(num1, num2):
    multiplicacao = float(num1) * float(num2)
    multiplicacao = str(multiplicacao).replace(".",",")
    print('A multiplica√ß√£o √© {}'.format(multiplicacao))

def Divisao (num1, num2):
    divisao = float(num1) / float(num2)
    divisao = str(divisao).replace(".",",")
    print('A divis√£o √© {}'.format(divisao))


def Entrada():
    num1 = input('Entre com o 1¬∞ n√∫mero: ')
    num1 = num1.replace(',','.')
    num2 = input('Entre com o 2¬∞ n√∫mero: ')
    num2 = num2.replace(',','.')
    return (num1, num2)

Entrada()
opcao= input('''Entre com o n√∫mero da op√ß√£o desejada:
1. Soma
2. Subtra√ß√£o
3. Multiplica√ß√£o
4. Divis√£o
''')

if opcao == '1':
    Soma(num1, num2)
elif opcao == '2':
    Subtracao(num1, num2)
elif opcao == '3':
    Multiplicacao(num1, num2)
elif opcao == '4':
   Divisao(num1, num2)
else:
    print('Erro, selecione uma op√ß√£o v√°lida!')
````
